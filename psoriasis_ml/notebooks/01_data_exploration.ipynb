# Notebook 01: Skin Lesion Data Exploration

This notebook contains the exploratory analysis of clinical skin photos. We will analyze the class balance (psoriasis vs. non-psoriasis), look at resolution spreads, and visualize sample lesions.

```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import collections

DATASET_PATH = "../dataset/train"
print(f"Checking dataset path: {DATASET_PATH}")
```

## 1. Analyze Class Distributions
Count the number of images in each sub-directory to verify class balance.

```python
if os.path.exists(DATASET_PATH):
    classes = [d for d in os.listdir(DATASET_PATH) if os.path.isdir(os.path.join(DATASET_PATH, d))]
    class_counts = {}
    for cls in classes:
        cls_dir = os.path.join(DATASET_PATH, cls)
        images = [f for f in os.listdir(cls_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        class_counts[cls] = len(images)
    
    print("Class Counts:", class_counts)
    
    # Plot the counts
    plt.figure(figsize=(6, 4))
    plt.bar(class_counts.keys(), class_counts.values(), color=['blue', 'orange'])
    plt.title("Skin Lesion Class Distribution")
    plt.ylabel("Number of Images")
    plt.show()
else:
    print("Dataset directory not found. Please upload images to Psoriasis_detection-main/dataset/train first.")
```

## 2. Visualize Sample Clinical Photos

```python
def show_samples(base_dir, num_samples=3):
    if not os.path.exists(base_dir): return
    classes = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]
    
    for cls in classes:
        cls_dir = os.path.join(base_dir, cls)
        images = [f for f in os.listdir(cls_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))][:num_samples]
        
        if len(images) == 0:
            continue
            
        fig, axes = plt.subplots(1, len(images), figsize=(12, 4))
        if len(images) == 1:
            axes = [axes]
        for i, img_name in enumerate(images):
            img_path = os.path.join(cls_dir, img_name)
            img = Image.open(img_path)
            axes[i].imshow(img)
            axes[i].set_title(f"{cls}: {img.size}")
            axes[i].axis('off')
        plt.suptitle(f"Sample images from {cls} class")
        plt.show()

show_samples(DATASET_PATH)
```
