## 2. Visualize Confusion Matrix
Renders a standard heatmap utilizing Seaborn for publication or thesis charts.

```python
def plot_confusion_matrix(cm, classes):
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=classes, yticklabels=classes)
    plt.title("Confusion Matrix - Skin Lesions")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    plt.show()

# Sample placeholder visualization
dummy_cm = np.array([[45, 5], [3, 47]])
plot_confusion_matrix(dummy_cm, ["psoriasis", "non_psoriasis"])
```
