# Notebook 02: Image Preprocessing Experimentation

In dermatological classification, lighting variance, scale differences, and shadows can significantly degrade classifier accuracy. This notebook allows you to test our custom autocontrast preprocessing pipeline on single test photos and observe the enhanced visual boundaries.

```python
import os
import sys
from PIL import Image, ImageOps
import matplotlib.pyplot as plt

# Add source path so we can import modules
sys.path.append(os.path.abspath('../src'))
from preprocessing import resize_and_normalize_image
```

## 1. Visualizing the effect of Autocontrast Normalization

```python
# Example visual test
test_image_path = "some_sample_lesion.jpg" # Replace with an actual file

if os.path.exists(test_image_path):
    orig = Image.open(test_image_path).convert('RGB')
    enhanced = ImageOps.autocontrast(orig)
    
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(orig)
    axes[0].set_title("Original Image")
    axes[0].axis('off')
    
    axes[1].imshow(enhanced)
    axes[1].set_title("After Autocontrast (Normalized Contrast)")
    axes[1].axis('off')
    
    plt.show()
else:
    print(f"To visualize, please place a sample image at: {test_image_path}")
```
